```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.neighbors import kneighbors_graph

# Generate 500 random data points
np.random.seed(123)
X = np.random.rand(500, 2)

# Create a KNN graph
k = 10 # number of neighbors for each data point
A = kneighbors_graph(X, k, mode='connectivity', include_self=True)

# Perform KMeans clustering using the graph Laplacian
n_clusters = 3 # number of clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=123)
kmeans.fit(A)

# Plot the data points and clusters
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
plt.title(f'KNN Clustering with K={k} and {n_clusters} Clusters')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

from sklearn.metrics import silhouette_score

# Calculate the silhouette score
silhouette_avg = silhouette_score(A, kmeans.labels_)
print(f'Silhouette Score: {silhouette_avg:.3f}')
```
