```python
import pandas as pd
data = pd. read_csv("/content/Bakery.csv")
data.head()
```

```python
data = data.drop(columns = ['DateTime', 'Daypart','DayType'])

data = data.drop_duplicates()

data. head()
```

```python
grouped_data = data.groupby('TransactionNo').aggregate(tuple)
grouped_data.head()
```

```python
transactions = grouped_data['Items'].tolist()
```

```python
itemsets, rules = apriori(transactions, min_support=0.01)
```

```python
for rule in rules:
  print (rule)
```
